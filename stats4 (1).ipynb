**Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

- **Probability Mass Function (PMF)**: PMF is a function that gives the probability that a discrete random variable is equal to a specific value. It maps each value of the random variable to its probability.

- **Probability Density Function (PDF)**: PDF is a function that describes the likelihood of a continuous random variable taking on a particular value. It gives the probability density at each point, but unlike the PMF, it doesn't directly give the probability of the variable being equal to a specific value due to the continuous nature of the variable.

**Example**: 
- For a fair six-sided die, the PMF would be: 
  - PMF(1) = 1/6, PMF(2) = 1/6, ..., PMF(6) = 1/6.
- For a continuous uniform distribution between 0 and 1, the PDF would be: 
  - PDF(x) = 1, for 0 <= x <= 1, and 0 otherwise.

**Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

- **Cumulative Density Function (CDF)**: CDF gives the probability that a random variable takes on a value less than or equal to a given value. It accumulates the probabilities up to a certain point.

**Example**: 
- For a fair six-sided die, the CDF would be: 
  - CDF(1) = 1/6, CDF(2) = 2/6, ..., CDF(6) = 1.

**Importance of CDF**: CDF provides a complete picture of the probability distribution of a random variable. It helps in understanding the probabilities associated with different values of the variable and is particularly useful in statistical analysis and hypothesis testing.

**Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.**

**Examples**: 
- Height and weight distributions in a population.
- IQ scores.
- Measurement errors.
- Stock prices changes.

**Parameters**:
- **Mean (μ)**: Determines the center of the distribution.
- **Standard Deviation (σ)**: Measures the spread or dispersion of the distribution.
- **Shape**: The normal distribution is symmetric and bell-shaped. The spread of the distribution is determined by the standard deviation. As the standard deviation increases, the distribution becomes wider.

**Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.**

**Importance**:
- Widely used in statistical analysis due to its mathematical tractability and properties.
- Central to hypothesis testing, confidence intervals, and estimation techniques.
- Acts as a benchmark for comparing other distributions.

**Examples**:
- Heights and weights of people in a population.
- IQ scores.
- Measurement errors.

**Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?**

**Bernoulli Distribution**: It models a single Bernoulli trial, which has only two possible outcomes: success (usually denoted by 1) and failure (usually denoted by 0), with a fixed probability of success (p).

**Example**: 
- Tossing a coin: Success (1) if it lands heads, failure (0) if it lands tails.

**Difference**:
- **Bernoulli Distribution**: Describes a single trial.
- **Binomial Distribution**: Describes the number of successes in a fixed number of independent Bernoulli trials.

**Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.**

**Solution**: 
- Using z-score formula: \( Z = \frac{x - \mu}{\sigma} \)
- \( Z = \frac{60 - 50}{10} = 1 \)
- Looking up the z-score table or using software, we find that the probability corresponding to a z-score of 1 is approximately 0.8413.
- Therefore, the probability that a randomly selected observation will be greater than 60 is approximately 0.8413.

**Q7: Explain uniform Distribution with an example.**

**Uniform Distribution**: It describes a continuous random variable where all intervals of the same length have an equal probability of occurring.

**Example**: 
- Rolling a fair six-sided die: Each outcome (1 to 6) has an equal probability of occurring, making it a uniform distribution.

**Q8: What is the z score? State the importance of the z score.**

**z-score**: It is a standardized score that indicates how many standard deviations a data point is from the mean. It's calculated as \( Z = \frac{x - \mu}{\sigma} \).

**Importance**: 
- Allows for comparison of values from different normal distributions.
- Provides information about how unusual or typical a data point is within a distribution.
- Useful in hypothesis testing and establishing confidence intervals.

**Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

**Central Limit Theorem (CLT)**: It states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution.

**Significance**:
- Allows for making inferences about population parameters from sample statistics.
- Forms the basis of many statistical techniques, including hypothesis testing and confidence intervals.

**Q10: State the assumptions of the Central Limit Theorem.**

**Assumptions**:
1. The sample is randomly selected from the population.
2. The sample size is large enough (usually n ≥ 30).
3. The population has a finite standard deviation.
